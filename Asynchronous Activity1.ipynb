```python
def print_numbers(fname):
    print("Telephone Numbers:")
    with open(fname,'r',encoding='utf-8') as f:
        data = f.read().split('\n')
    for namenum in data:
        if ',' in namenumm:
            namef , numf = namenum.split(',')
            print('Name:',namef,'\tNumber:',numf)
    print()
def add_number(f_name,name,number):
    with open(fname,'a',encoding='utf-8') as f:
        f.write(name + ',' + number + '\n')
        
def lookup_number(fname,name):
    with open(fname,'r',encoding='utf-8') as f:
        data = f.read().split('\n')
    numbers = {}
    for namenum in data:
        if ',' in namenum:
            namef , numf = namenum.split(',')
            numbers[namef] = numf
        
        if name in numbers:
            return "The number is " + numbers[name]
        else:
            return name + "was not found"
def print_menu():
    print("1. Print Phone no")
    print("2. Add a Phone no")
    print("3. Lookup a Phone no")
    print("4. Quit")
    
filename = 'numbers.txt'
while True:
    print_menu()
    menu_choice = int(input("Type in a no(1-4): "))
    if menu_choice == 1:
        print_numbers(filename)
    elif menu_choice == 2:
        print("Add Name and Number")
        name = input("Name: ")
        phone = input("Number: ")
        add_number(filename,name,phone)
    elif menu_choice == 3:
        print("Lookup number")
        name = input("Name: ")
        print(lookup_number(filename,name))
    elif menu_choice == 4:
        break
    else:
        print("Invalid choice.\n")
print("GoodBye")
```
