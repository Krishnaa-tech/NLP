### Extracting the text from txt file

```python
import spacy
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Load the English NLP model from spaCy
nlp = spacy.load('en_core_web_sm')

# Read text from a file
with open('document.txt', 'r', encoding='utf-8') as f:
    lines = f.read().strip()

#Split the text into paragraphs or lines (adjust as needed)
lines = lines.split('\n')

#Initialize a list to store clean tokens
clean_tokens = []

#Initialize a list to store cleaned text
cleaned_text_list = []

#Define symbols to be removed as they are extra strings in the corpos
symbols = "!\"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"


for line in lines:
    doc = nlp(line) # Tokenizing each line with spaCy
    
    for token in doc:
       
        clean_token = token.text.lower().strip() #Removes prefix, suffix, and make lowercase in the clean document
        
        #Removes numbers, symbols, URLs, and spaCy's stop words from the corpos
        if not clean_token.isdigit() and not any(char in clean_token for char in symbols) and not clean_token.startswith("http") and not token.is_stop:
            clean_tokens.append(clean_token)
    
    #Joining together the cleaned tokens to get the cleaned text
    cleaned_line = " ".join(clean_tokens)
    cleaned_text_list.append(cleaned_line)
    
    clean_tokens = []


cleaned_text = "\n".join(cleaned_text_list)


# Generate a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(cleaned_text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud")
plt.show()

print("******************************************")
# Print the cleaned text
print("Cleaned Text:", cleaned_text)
```
