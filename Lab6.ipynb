>## <center>Indian Language Stemmer

```python
import codecs
import re

#read the input file
filepath = 'hindi.txt'
f = codecs.open(filepath, encoding = 'utf-8')
text = f.read()


sentences=text.split(u"।") #since hindi sentences end with '|'
words_list = list()
for sentence in sentences:
    words = sentence.split(' ') #words are seperated by a space in hindi
    words_list += words

suffixes = {
    1: [u"ो",u"े",u"ू",u"ु",u"ी",u"ि",u"ा"],
    2: [u"कर",u"ाओ",u"िए",u"ाई",u"ाए",u"ने",u"नी",u"ना",u"ते",u"ीं",u"ती",u"ता",u"ाँ",u"ां",u"ों",u"ें"],
    3: [u"ाकर",u"ाइए",u"ाईं",u"ाया",u"ेगी",u"ेगा",u"ोगी",u"ोगे",u"ाने",u"ाना",u"ाते",u"ाती",u"ाता",u"तीं",u"ाओं",u"ाएं",u"ुओं",u"ुएं",u"ुआं"],
    4: [u"ाएगी",u"ाएगा",u"ाओगी",u"ाओगे",u"एंगी",u"ेंगी",u"एंगे",u"ेंगे",u"ूंगी",u"ूंगा",u"ातीं",u"नाओं",u"नाएं",u"ताओं",u"ताएं",u"ियाँ",u"ियों",u"ियां"],
    5: [u"ाएंगी",u"ाएंगे",u"ाऊंगी",u"ाऊंगा",u"ाइयाँ",u"ाइयों",u"ाइयां"],
} #suffixes tin hindi language that change tense, gender, number but not the meaning

stems=list()
for word in words_list:
    for L in range(1,5): 
        if len(word) >= L + 1:
            for suffix in suffixes[L]:
                if word.endswith(suffix):
                    word=word[:-L] #stripping the suffix from the word
    if word:
        stems.append(word)

filename='stems_generated.txt'
f=codecs.open(filename,"w",encoding='utf-8') #open in write mode
for stem in stems:
    f.write(str(stem))
    f.write(u"\u0020")
f.close()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import codecs

# Read the stemmed words from the text file
stemmed_words = []
with codecs.open('stems_generated.txt', encoding='utf-8') as file:
    for line in file:
        words = line.split()
        stemmed_words.extend(words)

# Print the stemmed words
print("Stemmed Words:")
for word in stemmed_words:
    print(word, end=' ')

# Join the stemmed words into a single string
stemmed_text = ' '.join(stemmed_words)

# Create a word cloud from the stemmed words
wordcloud = WordCloud(width=1000, height=700,
                      background_color='white',
                      min_font_size=10, font_path='gargi.ttf').generate(stemmed_text)

# Display the word cloud
plt.figure(figsize=(18, 8), facecolor=None)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()

```

It's a basic implementation of stemming, a common natural language processing technique used to reduce words to their root form for text analysis and language processing tasks.

- Here we declare the hindi stop words as suffixes and then we run the loop to match it with each word and then remove the suffixes from them nd then saves into a file named `stems_generated.txt` .

>## <center>Foreign Language Stemmer

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import SnowballStemmer
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Load the French stopwords
french_stopwords = set(stopwords.words('french'))

# Initialize the Snowball stemmer for French
stemmer = SnowballStemmer('french')

# Read text from a TXT file (replace 'your_file.txt' with the path to your file)
with open('french.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# Tokenize the text
words = word_tokenize(text, language='french')

# Remove stopwords and non-alphabetic words
words = [word.lower() for word in words if word.isalpha() and word.lower() not in french_stopwords]

# Perform stemming
stemmed_words = [stemmer.stem(word) for word in words]

# Save the stemmed words to a file
with open('stemmed_words.txt', 'w', encoding='utf-8') as stem_file:
    for word in stemmed_words:
        stem_file.write(word + '\n')

# Calculate word frequencies
word_freq = Counter(stemmed_words)

# Create a word cloud
wordcloud = WordCloud(width=1000, height=700, background_color='white').generate_from_frequencies(word_freq)

# Print the most common stemmed words and their frequencies
most_common = word_freq.most_common(10)  # Change the number to get more or fewer words
for word, freq in most_common:
    print(f'{word}: {freq}')
    
# Display the word cloud
plt.figure(figsize=(12, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title('Word Cloud of Stemmed French Words')
plt.show()
```

- Here we use `snowballstemmer()` to stem french words and save them to stemmed words and then obtain the word cloud stemming
