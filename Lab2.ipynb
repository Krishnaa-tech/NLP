```python
from PyPDF2 import PdfReader

f = open('document.pdf', 'rb')
reader = PdfReader(f)

text = ''

for page_number in range(len(reader.pages)):
    page = reader.pages[page_number]
    text += page.extract_text()

f.close()
```

```python
import spacy
from collections import Counter
import matplotlib.pyplot as plt

nlp = spacy.load('en_core_web_sm')

doc = nlp(text)

#Initialize a Counter to count word frequencies
word_freq = Counter()


stop_words = set(nlp.Defaults.stop_words)

#Iterate through each token in the spaCy document
for token in doc:
    clean_token = token.text.lower().strip()  # Convert to lowercase and strip whitespace

    # Check if the clean_token is alphanumeric and not a stop word
    if clean_token.isalpha() and clean_token not in stop_words:
        word_freq[clean_token] += 1

# Get the most common words and their frequencies
most_common_words = word_freq.most_common(10)  #Change the number to display more or fewer words

# Separate words and frequencies for plotting
words, frequencies = zip(*most_common_words)

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(words, frequencies)
plt.xlabel('Words')
plt.ylabel('Frequencies')
plt.title('Top Words by Frequency (Stop Words Removed)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()

# Print word frequencies
print("Word Frequency:")
for word, freq in word_freq.items():
    print(f"{word}: {freq}")
```
